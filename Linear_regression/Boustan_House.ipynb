```python
## i have to predict medv and remove those column which are not highly related to medv
```

```python
df = df.drop(["B", "CHAS","ZN","DIS","RAD","CRIM"], axis=1)
```

```python
df.head()
```

```python
df.corr()
```

```python
#visualize more closely
sns.pairplot(df)
```

```python
#independent and dependent feature
X = df[["INDUS", "NOX", "RM", "AGE", "TAX", "PTRATIO", "LSTAT"]]
Y=df['MEDV']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
 from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)#use for train only 
```

```python
X_test=scaler.transform(X_test)
```

```python
#apply linear regression 
from sklearn.linear_model import LinearRegression 
```

```python
regression=LinearRegression(n_jobs=-1)
```

```python
regression.fit(X_train,Y_train)
```

```python
print("Coefficient or slope beta1",regression.coef_)
print("Intercept",regression.intercept_)
```

```python
# select RM feature (index 2 → number of rooms)
# we choose RM because it has strong correlation with MEDV (≈ +0.69)
# meaning: as number of rooms increases, house price increases
X_rm = X_train[:, 2]
# plot actual data points (RM vs MEDV)
# this shows real relationship between rooms and price
plt.scatter(X_rm, Y_train)
# sort values to draw smooth best-fit line
import numpy as np
sorted_index = np.argsort(X_rm)
# arrange X and predicted values in increasing order
X_sorted = X_rm[sorted_index]
y_pred_sorted = regression.predict(X_train)[sorted_index]
# plot best-fit line (model prediction)
plt.plot(X_sorted, y_pred_sorted)
# label for x-axis (feature)
plt.xlabel("RM")
# label for y-axis (target)
plt.ylabel("MEDV")
# display graph
plt.show()
```

```python
#prediction for test data 
Y_pred=regression.predict(X_test)
```

```python
#performance metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mse=mean_squared_error(Y_test,Y_pred)
mae=mean_absolute_error(Y_test,Y_pred)
rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
print(mse)
print(mae)
print(rmse)
```

```python
#accuracy by r2
from sklearn.metrics import r2_score
```

```python
score=r2_score(Y_test,Y_pred)
```

```python
print(score)
```

```python
#adjusted r2
1-(1-score)*(len(Y_test)-1)/(len(Y_test)-X_test.shape[1]-1)
```

```python
#prediction for new data 
# first scale input using same scaler
new_data = [[5.0, 0.5, 6.5, 60, 300, 15, 12]]
scaled_input = scaler.transform(new_data)
prediction = regression.predict(scaled_input)
print(prediction)
```
